```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
gender_counts = df['Region'].value_counts()
bar_width = 0.9
x=range(len(gender_counts.index))

plt.bar(gender_counts.index,gender_counts.values)
plt.xlabel('Region')
plt.ylabel('Count')
plt.title('Distribution of Region')

plt.xticks(x,gender_counts.index,rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['Region'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Population in 2020')
plt.xlabel('Total Population')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
